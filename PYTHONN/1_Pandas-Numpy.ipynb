### **1. Analyzing Weather Data**

#### **Write a NumPy program to analyze a set of temperature data (in degrees Celsius) for a city. Calculate the average temperature, the variance, and identify the days with temperatures above the 90th percentile**.

**Input:**

[22, 24, 28, 32, 31, 30, 35, 36, 38, 40, 41, 42, 43, 44, 45]


**Output:**

Average Temperature: 35.4

Variance: 50.11

Days above 90th percentile: [44, 45]


**Input:**

[25, 27, 29, 31, 33, 35, 37, 39, 21, 23, 25, 27, 29, 31, 33]


**Output:**

Average Temperature: 29.67

Variance: 24.89

Days above 90th percentile: [37, 39]


````python
# Write your code here
import numpy as np

def analyze_temperatures(temps):
    temperatures = np.array(temps)
    average_temp = np.round(np.mean(temperatures),2) # Rounding off to 2 decimal places
    variance = np.round(np.var(temperatures), 2)
    percentile_90th = np.percentile(temperatures, 90)
    days_above_percentile = temperatures[temperatures > percentile_90th].tolist()

    return average_temp, variance, days_above_percentile

# Reading input dynamically
temps = eval(input())

# Calling the function and displaying the result
result = analyze_temperatures(temps)
print("Average Temperature:", result[0])
print("Variance:", result[1])
print("Days above 90th percentile:", result[2])
````

### **2. Comparing Stock Prices**
#### **Create a NumPy program to compare the daily closing prices of two stocks over a month. Calculate the days on which stock A's price was greater than stock B's, and compute the average price difference on those days.**

**Input:**

Stock A: [220, 225, 210, 215, 220, 230, 235, 240, 245, 250]

Stock B: [200, 210, 205, 215, 220, 225, 230, 240, 235, 245]

**Output:**

Days A > B: [0, 1, 2, 5, 6, 8, 9]

Average Price Difference: 9.29


**Input:**

Stock A: [310, 320, 315, 310, 305, 300, 295, 290, 285, 280]

Stock B: [300, 305, 310, 315, 320, 325, 330, 335, 340, 345]

**Output:**

Days A > B: [0, 1, 2]

Average Price Difference: 10.0

````python
import numpy as np

def compare_stock_prices(stock_a, stock_b):
    a_prices = np.array(stock_a)
    b_prices = np.array(stock_b)
    days_a_greater_than_b = np.where(a_prices > b_prices)[0].tolist()
    avg_diff = np.round(np.mean(a_prices[days_a_greater_than_b] - b_prices[days_a_greater_than_b]),2)

    return days_a_greater_than_b, avg_diff

# Reading input dynamically
stock_a = eval(input("Stock A:"))
stock_b = eval(input("Stock B:"))

# Calling the function and displaying the result
result = compare_stock_prices(stock_a, stock_b)
print("Days A > B:", result[0])
print("Average Price Difference:", result[1])
````

### **3. Matrix Operation on Sensor Data**
#### **Given a 4x4 matrix representing sensor readings over four days, normalize the data by subtracting the mean and dividing by the standard deviation for each day. Then, calculate the sum of the normalized values for each sensor.**

**Input:**

[[10, 20, 30, 40],

 [15, 25, 35, 45],

 [20, 30, 40, 50],

 [25, 35, 45, 55]]

 **Output:**

Normalized Matrix:

[-1.3416407864998738, -0.4472135954999579, 0.4472135954999579, 1.3416407864998738]

[-1.3416407864998738, -0.4472135954999579, 0.4472135954999579, 1.3416407864998738]

[-1.3416407864998738, -0.4472135954999579, 0.4472135954999579, 1.3416407864998738]

[-1.3416407864998738, -0.4472135954999579, 0.4472135954999579, 1.3416407864998738]

Sum per Sensor: [-5.366563145999495, -1.7888543819998317, 1.7888543819998317, 5.366563145999495]

````python
import numpy as np

def normalize_sensor_data(matrix):
    data = np.array(matrix)
    normalized_data = (data - np.mean(data, axis=1, keepdims=True)) / np.std(data, axis=1, keepdims=True)
    sum_per_sensor = np.sum(normalized_data, axis=0).tolist()

    return normalized_data.tolist(), sum_per_sensor

# Reading input dynamically
matrix = eval(input("Enter a 4x4 matrix: "))

# Calling the function and displaying the result
normalized_matrix, sum_per_sensor = normalize_sensor_data(matrix)
print("Normalized Matrix:")
for row in normalized_matrix:
    print(row)
print("Sum per Sensor:", sum_per_sensor)
````

### **4. Distance Between Points**
#### **Write a NumPy program to calculate the Euclidean distance between two sets of points in 3D space**.

**Input:**

[[1, 2, 3], [4, 5, 6]]

[[7, 8, 9], [1, 3, 5]]

**Output:**

[10.392, 3.742]

**Input:**

[[0, 1, 2], [3, 4, 5]]

[[6, 7, 8], [0, 2, 4]]

**Output:**

[10.392,   3.742]

Screenshot 2024-01-04 at 5.07.38â€¯PM.png

````python
import numpy as np

def euclidean_distance(points_a, points_b):
    a = np.array(points_a)
    b = np.array(points_b)
    distance = np.sqrt(np.sum((a - b)**2))
    return np.round(distance, 3).tolist()

# Reading input dynamically
points_a = eval(input(""))
points_b = eval(input(""))

# Calling the function and displaying the result
print(euclidean_distance(points_a, points_b))
````

### **7. Flight Path Analysis**
#### **Develop a NumPy program to analyze flight path data. Given an array of altitude readings (in meters) over time, compute the total ascent and descent during the flight. Assume the flight starts and ends at the same altitude.**

**Input:**

[100, 150, 200, 150, 100, 50, 100, 150]

**Output**:

Total Ascent: 200

Total Descent: 150


**Input:**

[100, 120, 180, 140, 160, 100]

**Output**:

Total Ascent: 100

Total Descent: 100

````python
import numpy as np

def flight_path_analysis(altitudes):
    altitudes_array = np.array(altitudes)
    differences = np.diff(altitudes_array)
    total_ascent = np.sum(differences[differences > 0])
    total_descent = np.sum(differences[differences < 0])
    return total_ascent, abs(total_descent)

# Reading input dynamically
altitudes = eval(input(""))

# Calling the function and displaying the result
ascent, descent = flight_path_analysis(altitudes)
print("Total Ascent:", ascent)
print("Total Descent:", descent)
````

#### **Segment customers into three groups based on 'Purchase_Amount': Low (<200), Medium (200-400), High (>400). Calculate the average purchase amount for each region and group.**

**Table: customers**



```
+-------------+-----------------+---------------+--------+
| Customer_ID | Purchase_Amount | Purchase_Date | Region |
+-------------+-----------------+---------------+--------+
| 1001        | 250             | 2023-02-15    | North  |
| 1002        | 500             | 2023-02-16    | South  |
| 1003        | 150             | 2023-02-17    | East   |
| 1004        | 450             | 2023-02-18    | West   |
| 1005        | 300             | 2023-02-19    | North  |
+-------------+-----------------+---------------+--------+

```



**Expected Output:**



```
+--------+---------+-------------------+
| Region | Segment | Average_Purchase  |
+--------+---------+-------------------+
| East   | Low     | 150.0             |
| North  | Medium  | 275.0             |
| South  | High    | 500.0             |
| West   | High    | 450.0             |
+--------+---------+-------------------+


```



````python
customers = pd.DataFrame({
    'Customer_ID': [1001, 1002, 1003, 1004, 1005],
    'Purchase_Amount': [250, 500, 150, 450, 300],
    'Purchase_Date': ['2023-02-15', '2023-02-16', '2023-02-17', '2023-02-18', '2023-02-19'],
    'Region': ['North', 'South', 'East', 'West', 'North']
})
````

````python
# Write your code here
def segment_group(amount):
    if amount < 200:
        return 'Low'
    elif 200 <= amount <= 400:
        return 'Medium'
    else:
        return 'High'

customers['Segment'] = customers['Purchase_Amount'].apply(lambda x: segment_group(x))
average_purchase = customers.groupby(['Region', 'Segment'])['Purchase_Amount'].mean().reset_index()

# Renaming the column for clarity
average_purchase.rename(columns={'Purchase_Amount': 'Average_Purchase'}, inplace=True)

# Displaying the modified DataFrame
print(average_purchase.to_string(index=False))
````

Solution in SQL:


```
SELECT Region,
       CASE
           WHEN Purchase_Amount < 200 THEN 'Low'
           WHEN Purchase_Amount BETWEEN 200 AND 400 THEN 'Medium'
           ELSE 'High'
       END AS Segment,
       AVG(Purchase_Amount) AS Average_Purchase  
FROM customers
GROUP BY Region, Segment;
```


The .idxmax() function in Pandas is used to find the index of the first occurrence of the maximum value over the requested axis. In simple terms, it helps you find the position of the highest value in a column (for a Series) or in each row/column (for a DataFrame) of a Pandas object.

Here's a bit more detail:

For a Series: If you apply .idxmax() to a Pandas Series, it returns the index label of the maximum value in the Series.

For a DataFrame: When used on a DataFrame, you can specify the axis along which to find the indices of the maximum values.


1. If axis=0 (default), it finds the index of the maximum value in each column.
2. If axis=1, it finds the index of the maximum value in each row.
