```python
# Roots of any polynomial equation
import numpy as np
# Define the coefficients of the polynomial equation
coefficients = [1, 0, 13, -12]  # coefficients of x^3, x^2, x, constant term
# Find the roots of the polynomial equation
roots = np.roots(coefficients)
print("Roots of the polynomial equation:", roots)
```

```python
# Combination of the numbers  unlike nested loops
from itertools import combinations
from itertools import permutations
for i in permutations([1,2,3],2):
  print(i)
```
