# 3. Variables

```python
# Static Vs Dynamic Typing
# Static Vs Dynamic Binding
# stylish declaration techniques
```

```python
# Typing decides the variable data type, like in static typing you've to specify the type prior unlike dynamic in which
# interpretor understands the variable data type.

# In static binding you cant change the variable data type while changing the values associated with it where as in dynamic binding you can
# change the variable values along with thier data types.
```

```python
# C/C++
name = 'nitish'
print(name)

a = 5
b = 6

print(a + b)
```

```python
# Dynamic Typing
a = 5                             #interpretor assigns the data types
# Static Typing
int a = 5
```

```python
# Dynamic Binding                #can assign new value to the interpretor
a = 5
print(a)
a = 'nitish'
print(a)

# Static Binding             # Static Binding/ where u can't change type of variable
int a = 5
```
