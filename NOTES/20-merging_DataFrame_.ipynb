```python
# IPL Problems
# find top 3 stadiums with highest sixes/match ratio
# find orange cap holder of all the seasons
```

```python
cc=delivery.merge(matches,right_on='id',left_on='match_id',how='inner')
c=cc.groupby('city',group_keys=False)
def ratio(df):
  count_six=df[df['batsman_runs']==6]['match_id'].count()
  count_matches=df['match_id'].nunique()
  df['ratio']=count_six/count_matches
  return df
c.apply(ratio)[['city','ratio']].groupby('city')['ratio'].mean().sort_values(ascending=False).head(3)
```
