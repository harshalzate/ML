1. Passes
2. Passes Completed
3. Attempted Line Breaks
4. Completed Line Breaks

You have to make a python function named `z_normalization` which takes two arguments:

1. *group:* Every group that you have created
2. *cols_to_perform:* This parameter takes a list of columns on which you have to perform the Z-Normalization.

For this problem, you have to use th `apply()` method.

$\Large Z \ - \ Normalization = \frac{X_i - \mu}{std} $

After that find out the below values for each group:
- minimum "Passess"
- maximum "Passess"
- minimum "Yellow Cards"
- maximum "Yellow cards"
- average "Yellow Cards"
- maximum "Attempted Line Breaks"
- minimum "Attempted Line Breaks"
- standardard deviation of "Attempted Line Breaks"
- average Possession

```python
df['Passes'].std()
```

```python
team=df.groupby('Team',group_keys=False)
```

```python
def norm(team):
  mean=team['Passes'].mean()
  std=team['Passes'].std()
  team['normalized_passes']=(team['Passes']-mean)/std
  return team
team.apply(norm)
```

```python
teams=df.groupby('Team',group_keys=False)   ## Group keys False is very important
l=['Passes','Passes Completed','Attempted Line Breaks','Completed Line Breaks']
def z_normalization(group,l):
  n=l
  for i in n:
    group['normalized_{}'.format(i)]=(group[i]-group[i].mean())/(group[i].std())
  return (group)                                                ## return a group of each group sent and eventually gets appended in a single dataframe
df1=teams.apply(z_normalization,l)
df1
```
