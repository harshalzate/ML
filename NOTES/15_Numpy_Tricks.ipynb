### np.where

The numpy.where() function returns the **indices** of elements in an input array where the given condition is satisfied when if else is not given.
On giving if else it return the elements based on the conditions

https://numpy.org/doc/stable/reference/generated/numpy.where.html

```python
a
```

```python
# find all indices with value greater than 50
np.where(a>50)
```

```python
np.where(a>0) # only way to get indexes
```

```python
np.where(a>50,9,a)
```

```python
np.where(a>50,a,0)
```

```python
# replace all values > 50 with 0
np.where(a>50,0,a)
```

```python
np.where(a%2 == 0,0,a)
```

### Set functions

- np.union1d

- np.in1d

```python
m = np.array([1,2,3,4,5])
n = np.array([3,4,5,6,7])
np.union1d(m,n)
```

```python
np.setdiff1d(n,m)
```

```python
m[np.in1d(m,1)]
```

```python
# Multi-dimensional arrays
arr3 = np.array([[1, 2], [3, 4]])
arr4 = np.array([[3, 4], [5, 6]])

# Intersection of arr3 and arr4 (result will be 1D)
intersection_multi = np.intersect1d(arr3, arr4)
print(intersection_multi)  # Output: [3 4]
```
