1. Straight Line

2. Square Pattern

Input: n (side length of the square)

Logic: Nested loop with both the outer and inner loop running n times each.

Formula: N/A

````python
for i in range(n):  # Rows
    for j in range(n):  # Columns
        print("*", end=" ")
    print()
````

3. Right-Angled Triangle

Input: n (height of the triangle)

Logic: Outer loop for rows runs n times; inner loop for columns runs up to the current row number.

Formula: Columns in row i = i + 1

````python
for i in range(n):  # Rows
    for j in range(i + 1):  # Columns
        print("*", end=" ")
    print()
````

4. Inverted Right-Angled Triangle

Input: n (height of the triangle)

Logic: Outer loop for rows runs n times; inner loop for columns runs from n to i (inclusive).

Formula: Columns in row i = n - i

````python
for i in range(n):  # Rows
    for j in range(n - i):  # Columns
        print("*", end=" ")
    print()
````

5. Pyramid Pattern

Input: n (height of the pyramid)

Logic: Outer loop for rows runs n times; two inner loops, one for spaces (n - i - 1) and one for asterisks (2 * i + 1).

Formula: Spaces before stars in row i = n - i - 1, Stars in row i = 2 * i + 1

````python
for i in range(n):  # Rows
    for j in range(n - i - 1):  # Spaces
        print(" ", end="")
    for j in range(2 * i + 1):  # Stars
        print("*", end="")
    print()
````

6. Diamond Pattern

Input: n (half of the height of the diamond)

Logic: Two parts; upper half is a pyramid, lower half is an inverted pyramid.

Formula: Upper half same as pyramid, Lower half inverted pyramid logic

````python
# Upper half
for i in range(n):
    print(" " * (n - i - 1) + "*" * (2 * i + 1))
# Lower half
for i in range(n - 2, -1, -1):
    print(" " * (n - i - 1) + "*" * (2 * i + 1))
````

7. Hollow Square

Input: n (side length of the square)

Logic: Nested loop; special condition for borders (first and last row/column).

Formula: N/A


````python
for i in range(n):  # Rows
    for j in range(n):  # Columns
        if i == 0 or i == n - 1 or j == 0 or j == n - 1:  # Border condition
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
````

8. Numeric Patterns

Input: n (height of the pattern)

Logic: Outer loop for rows; inner loop for numbers, often with incrementing values.

Formula: Varies; often involves simple arithmetic progression

````python
for i in range(n):  # Rows
    num = 1
    for j in range(n - i - 1):  # Spaces
        print(" ", end="")
    for j in range(i + 1):  # Numbers
        print(num, end=" ")
        num += 1
    print()
````

10. Checkerboard Pattern

Input: n (size of the checkerboard)

Logic: Nested loop with a condition to alternate printing spaces and asterisks.

Formula: Print '*' if (i + j) % 2 == 0 else print space.

````python
for i in range(n):
    for j in range(n):
        if (i + j) % 2 == 0:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
````

11. Hollow Pyramid

Input: n (height of the pyramid)

Logic: Use nested loops to create a pyramid. Print asterisks only on the borders and spaces inside.

Formula: Print '*' at start/end of row and at pyramid borders, else print space.

````python
for i in range(n):
    for j in range(n - i - 1):
        print(" ", end="")
    for j in range(2 * i + 1):
        if j == 0 or j == 2 * i or i == n - 1:
            print("*", end="")
        else:
            print(" ", end="")
    print()
````

12. Zig-Zag Pattern

Input: n (number of cycles), m (length of each cycle)

Logic: Loop through the length of the pattern, using conditions to print asterisks in a zig-zag fashion.

Formula: Print '*' if (row + column) % (m - 1) == 0 or row == column % (m - 1).

````python
m=5
for i in range(n):
    for j in range(m * n):
        if (i + j) % (m - 1) == 0 or (i == j % (m - 1)):
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
````

13. Fibonacci Pattern

Input: n (number of rows)

Logic: Generate Fibonacci series and print the numbers in a pattern.

Formula: fib(n) = fib(n-1) + fib(n-2), with fib(0) = 0 and fib(1) = 1.

````python
n=5
def fibonacci(n):
    a, b = 0, 1
    for i in range(n):
        for j in range(i + 1):
            print(a, end=" ")
            a, b = b, a + b
        print()


fibonacci(n)
````

14. Wave Pattern

Input: n (number of rows), m (number of waves)

Logic: Nested loop with a condition to create wave-like structure using asterisks.

Formula: Print '*' if row is equal to column % 4 or 4 - column % 4, else print space.

````python
for row in range(n):
    for column in range(m * 4):
        if row == column % 4 or row == 4 - column % 4:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
````

15. Diamond Inside Square

Input: n (side length of the square and half the diagonal of the diamond)

Logic: Use nested loops and conditions to print a diamond pattern inside a square.

Formula: Print '*' for diamond and square borders; use conditions to differentiate.

````python
for i in range(2 * n):
    for j in range(2 * n):
        if i == 0 or i == 2 * n - 1 or j == 0 or j == 2 * n - 1 \
        or i == j or i + j == 2 * n - 1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
````

16. Hourglass Pattern

Input: n (half height of the hourglass)

Logic: Two parts; an inverted pyramid on top and a straight pyramid at the bottom.

Formula: Use loops to create the upper and lower halves with decreasing and increasing stars, respectively.

````python
for i in range(n, 0, -1):
    print(" " * (n - i) + "* " * i)
for i in range(2, n + 1):
    print(" " * (n - i) + "* " * i)
````

17. Butterfly Pattern

Input: n (half of the height of the butterfly)

Logic: Two parts; mirrored right-angled triangles for upper and lower wings.

Formula: Print asterisks for the triangle parts and spaces in between.

````python
for i in range(n):
    print("*" * (i + 1) + " " * (2 * (n - i - 1)) + "*" * (i + 1))
for i in range(n - 1, -1, -1):
    print("*" * (i + 1) + " " * (2 * (n - i - 1)) + "*" * (i + 1))
````

18. Cross Pattern

Input: n (dimension of the cross, better if it's an odd number)

Logic: Use conditionals within loops to print a cross.

Formula: Print '*' for the middle row and column; otherwise, print space.

````python
for i in range(n):
    for j in range(n):
        if i == n // 2 or j == n // 2:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
````

19. Circle Pattern

Input: radius (radius of the circle)

Logic: Use nested loops to iterate through a grid, printing characters based on their distance from the center of the circle.

Formula: Print '*' if the distance sqrt((i - radius) ** 2 + (j - radius) ** 2) from the center is less than or equal to radius.

````python
from math import sqrt
radius = 5
for i in range(2 * radius + 1):
    for j in range(2 * radius + 1):
        if sqrt((i - radius) ** 2 + (j - radius) ** 2) <= radius:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
````

20. Hollow Circle Pattern

Input: radius (radius of the hollow circle)

Logic: Similar to the circle pattern, but only print characters if they are near the circumference.

Formula: Print '*' if the distance from the center is within a small range of the radius.

````python
from math import sqrt
for i in range(2 * radius + 1):
    for j in range(2 * radius + 1):
        distance = sqrt((i - radius) ** 2 + (j - radius) ** 2)
        if radius - 0.5 < distance < radius + 0.5:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
````

21. Pascal's Triangle

Input: n (number of rows in the triangle)

Logic: Use loop structure to calculate values of Pascal’s triangle using the formula C(line, i) = line! / ((line - i)! * i!).

Formula: Each number is the sum of the two directly above it.

````python
from math import factorial
for line in range(1, n + 1):
    for i in range(n - line + 1):
        print(" ", end="")
    for i in range(line):
        print(factorial(line - 1) // (factorial(line - 1 - i) * factorial(i)), end=" ")
    print()
````

22. Inverted Pascal's Triangle

Input: n (number of rows in the triangle)

Logic: Similar to Pascal's triangle but inverted. Start from the nth line and decrease.

Formula: Each number is the sum of the two directly above it in Pascal's triangle.

````python
from math import factorial
for line in range(n, 0, -1):
    for i in range(n - line + 1):
        print(" ", end="")
    for i in range(line):
        print(factorial(line - 1) // (factorial(line - 1 - i) * factorial(i)), end=" ")
    print()
````
