```python
# is a function provided by the ast (Abstract Syntax Trees) module.
# It is designed specifically for safely evaluating Python literals (constants) and container displays (e.g., lists, dictionaries, tuples).
import ast
# Example string containing a Python literal (list)
literal_string = '''['Power Back-up', 'Intercom Facility', 'Lift(s)', 'Swimming Pool', 'Park', 'Piped-gas',
              'Internet/wi-fi connectivity', 'Fitness Centre / GYM', 'Club house / Community Center']'''
# Using ast.literal_eval to safely evaluate the string
result_list = ast.literal_eval(literal_string)
print(result_list)
```

```python
def drawDiamondPattern(size):
    for i in range(size):
        print(' '*(size-i-1),end='')
        for j in range((i*2) +1):
            print(j+1,end='')
        print(' '*(size-i-1),end='')
        print('')
    for k in range(size-1,0,-1):
        print(' '*(size-k),end='')
        for l in range(2*(k) -1):
            print(l+1,end='')
        print(' '*(size-k),end='')
        print('')

size = int(input())

# Call the function to draw the diamond pattern
drawDiamondPattern(size)
```
