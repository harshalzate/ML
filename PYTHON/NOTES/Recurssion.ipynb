```python
# Multiplication
def rec(a,b):
  if b==1: # 1. Base condition
    return a
  else:    # 2. Logic
    return a*1 + rec(a,b-1)
rec(4,3)
```

```python
# Multiplication
def mul(a,b):
  if b==0:
    return 0
  else:
    return a + mul(a,b-1)
mul(3,99)
```

```python
# Table
def table(n,current=1):   # Assigning variables inititally only like counter
  if current>10:
    return
  else:
    print(n*current)
    table(n,current=current+1)
table(5)
```

```python
# Table
def table(n,counter=1):
  if counter==10:
    return print(n*counter)
  else:
    print(n*counter)
    counter = counter+ 1
    table(n,counter)

table(20)
```

```python
# Factorial
def fact(n):
  if n==1 or n==0:  # base condition to terminate the loop
    return 1
  return n * fact(n-1)  # decomposition logic
fact(5)
```

```python
# Palindrome
def palindrome(string):
  if len(string)<=1:
    print( "palindrome")
  else:
    if string[0]==string[-1]:
      palindrome(string[1:-1])
    else:
      print("Not a Palindrome")
palindrome('nayan')
```

```python
# fibonacci
def fibonacci(n):
    if n <= 1:
        return n
    else:
        return fibonacci(n-1) + fibonacci(n-2)

# Example usage
for i in range(4):
    print(f"Fibonacci({i}) = {fibonacci(i)}")
```

```python
# fibonacci
def fibonacci(n,curr=1,a=0,b=1):
  if curr==n:
    return
  else:
    c=a+b
    print(c)
    a=b
    b=c
    n=n-1
    fibonacci(n,curr,a,b)
fibonacci(5)
```

```python
# Fibonacci Memorization    ###
def fibonacci(n):           ### stores the values of the already calculated functions and then moves forward if the values are not calculated already
  d={1:0,2:1}
  if n in d.keys():
    return d[n]
  else:
    d[n]=fibonacci(n-1) + fibonacci(n-2)
    return d[n]
fibonacci(10)
```

```python
def factorial(n):
  if n==1:
    return 1
  else:
    return n*factorial(n-1)
factorial(5)
```
