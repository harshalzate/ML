```python
from sklearn.datasets import make_classification
import numpy as np

# Create an imbalanced dataset
X, y = make_classification(n_samples=1000,    # Number of samples
                           n_features=20,     # Number of features
                           n_informative=2,   # Number of informative features
                           n_redundant=10,    # Number of redundant features
                           n_clusters_per_class=1,
                           weights=[0.75],    # Proportion of the majority class
                           flip_y=0,          # Fraction of samples whose class is assigned randomly
                           random_state=42)   # Random seed

# Check the distribution of the target variable
unique, counts = np.unique(y, return_counts=True)
print(f"Class distribution: {dict(zip(unique, counts))}")
```

```python
# Used to judge the performance of algorithm on the dataset with generllization and without any bias
```

```python
# leave one out
from sklearn.model_selection import LeaveOneOut
lr=LogisticRegression()
loo=LeaveOneOut()
scores=cross_val_score(lr,X,y,cv=loo,scoring='accuracy')
print(np.mean(scores))  # mean of accuracy for the classification and mean of regression metric for the regression
print(np.std(scores))
```

```python
# KFold
from sklearn.model_selection import KFold
kfold=KFold(n_splits=5,shuffle=True)
print(np.mean(cross_val_score(lr,X,y,cv=kfold,scoring='accuracy')))
print(np.std(cross_val_score(lr,X,y,cv=kfold,scoring='accuracy')))
```

```python
# stratified k fold
from sklearn.model_selection import StratifiedKFold
skf=StratifiedKFold(n_splits=5,shuffle=True)
print(np.mean(cross_val_score(lr,X,y,cv=skf,scoring='accuracy')))
print(np.std(cross_val_score(lr,X,y,cv=skf,scoring='accuracy')))
```

```python
# by default cv= kfold and takes n=split
from sklearn.model_selection import KFold
cv=cross_val_score(lr,X,y,cv=3)
print(np.mean(cv))
print(np.std(cv))
```
