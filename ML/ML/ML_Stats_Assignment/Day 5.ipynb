## 1. What is Semi- Supervised Learning and why is it so important?


Semi-supervised learning is a machine learning technique that bridges the gap between supervised and unsupervised learning. Here's the breakdown:

Supervised learning requires a lot of labeled data, where each data point has a corresponding answer or label. For example, training an image classifier with labeled pictures of cats and dogs.
Unsupervised learning deals with unlabeled data, where the algorithm must find patterns or structures on its own. For instance, grouping similar customer purchases into clusters.
Semi-supervised learning leverages the strengths of both approaches. It uses a small amount of labeled data combined with a large amount of unlabeled data to train a model. This is important for several reasons:

Data Labeling Cost: Labeling data can be expensive and time-consuming, especially for complex tasks. Semi-supervised learning allows you to get good results with less labeled data, reducing costs.
Abundance of Unlabeled Data: Many situations have vast amounts of unlabeled data readily available. Social media text, sensor readings, or website clickstream data are all examples. Semi-supervised learning lets you harness this unlabeled data to improve your model.
Improved Performance: By incorporating the unlabeled data, semi-supervised learning algorithms can often achieve better performance compared to using only labeled data.
Here's an analogy: Imagine you're learning a new language. Supervised learning would be like having a teacher give you labeled flashcards (word on one side, definition on the other). Unsupervised learning would be like being thrown into a conversation with native speakers and trying to pick up the language by ear. Semi-supervised learning is like having a few flashcards to get you started, then immersing yourself in conversations to solidify your understanding.

## 2. What is stratified cross-validation and when should we use it?

Stratified cross-validation is a technique used to evaluate the performance of a machine learning model, particularly in situations where the distribution of classes (or categories) in the dataset is imbalanced. It is an extension of k-fold cross-validation, where the dataset is divided into k subsets (folds), and the model is trained and evaluated k times, each time using a different fold as the validation set and the remaining folds as the training set.

In stratified cross-validation:

The dataset is divided into k folds while preserving the percentage of samples for each class in each fold.
During each iteration of cross-validation, the model is trained on k-1 folds and evaluated on the remaining fold. This process is repeated k times, with each fold used once as the validation set.
By ensuring that each fold maintains the same class distribution as the original dataset, stratified cross-validation helps in producing more reliable and representative estimates of the model's performance, especially when dealing with imbalanced datasets.

## 3. What do you understand about Information Gain? Also, explain the mathematical formulation associated with it.



Information Gain is a concept commonly used in decision trees and other machine learning algorithms for feature selection. It quantifies the amount of uncertainty (or randomness) in a dataset that is reduced when a particular feature is used for splitting the data.

In simple terms, Information Gain measures how much knowledge a feature provides about the target variable. Features that result in higher information gain are considered more useful for splitting the dataset because they help distinguish between different classes more effectively.

Mathematically, Information Gain is calculated using the entropy measure. Entropy is a measure of randomness or uncertainty in a dataset.
Information Gain (IG) for a feature with respect to a dataset is then calculated as the difference between the entropy of the original dataset and the weighted sum of entropies after splitting the dataset based on feature
