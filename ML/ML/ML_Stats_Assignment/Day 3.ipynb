## 1. What do you understand by Pruning in a Decision Tree and Why do we require Pruning in Decision Trees? Explain.

A technique used to reduce or restrict the size of the decision tree in order to avoid overfitting.
Ther are two types of pruning methods:
1. Pre-pruning: This halt the tree construction early and doesn't grow the entire tree construction, done in various ways to set the limit on the depth of the tree.

a. Maximum depth: set a limit on maximum depth of the tree i.e maximum node splitings.

b. maximum sample_splits: If the samples in any of the node are below certain threshold then don't split it .

c. maximum samplt_leaf: If the samples in any of the leaf are below certain threshold after a split then don't split it anymore.

d. Maximum leaf nodes: Limit the no of nodes which are allowed to form.

e. Maximum Impurity decrease: Works on the similar principle as decision tree only, if the reduction in impuruty is not above certain threshold after a split, you stop the splitting.

Advanteages:

I. Simplicity

II. Computational efficiency

Disadvantages:

I. Risk of underfitting

II. Require fine tuning

III. short-sightdeness : It can neglect good nodes (in terms of impurity reduction) cause of early pruning which would have arrived after the node where pruning was done.


2. Post-pruning: Allows a tree to grow completely and then prun those nodes which are least significant in terms of impurity reduction

## 3. What is Model Explainability and why is it required?


What is model explainibility

- Model explainability is the ability to understand the decisions made by the
model. So, it's inferential ability of the model to explain the decision and result. In certain application like medicare, finicial frauds it is crucial to know the factors which leads you to that decision so that those could be avoided and rectified in these applications inference ability of the model plays crucial role aprt from the obivious prediction ability.

- Model explainability is usually done by feature importance, so the feature with high feature importance weight is more significant contributing factor for the outcome. Tree based models like decision trees, random forest and gradient boosting gives us feature importances for each feature, likewise linear model like linear regression, logistic regression also gives us weighted coefficents which are treated as importance of that feature, in both the above cases the greater the feature importance or coefficent value greater the importance of that feature which explain the fact that this much of the prediction outcome causes are explain by this feauture.  In linear regression a regression analysis is done to better understand the impact of feature and it's coefficents on the indepndent varible

- Trust and Transparency: In many fields, like healthcare or finance, trusting the model's decisions is crucial. Explainability helps users understand why the model made a certain prediction, which builds trust and allows for better decision-making.

- Avoiding Bias and Errors: By understanding how the model works, we can identify potential biases in the training data that might skew its results. Explainability helps us catch errors and ensure the model is making fair and accurate predictions.

- Imagine a model that predicts loan approvals. Without explainability, you wouldn't know why someone's loan was rejected. Explainability could reveal hidden biases or errors, ensuring fair decisions for everyone.

Here are some additional benefits of explainability:

- Improved Model Development: By understanding how the model works, developers can identify areas for improvement and fine-tune its performance.
Regulatory Compliance: In some industries, regulations require models to be explainable for auditing purposes.
There are different techniques for achieving explainability, depending on the type of model. But the core idea is to make the inner workings of the model more transparent and understandable to humans.
