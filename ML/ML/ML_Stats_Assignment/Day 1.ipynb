## 3. Briefly explain Hyperparameter Optimization strategies and optimization strategy performance.

Hyperparameter Optimzaton is the proecsse of optimizing the hyperparameters which are not learned during model learning phase but they control the aspect of learning only. We have to do trial and error to get the best hyperparamets and strategies to optimize this are:
1. Grid search CV: Grid search performs a exhaustive search over a specific grid of combination of hyperparameters where each point on the grid represent a unique combination of the hyperparametes. Like this we calculate for each combination and select the best model, a cross validation technique is incorporated in the grid search cv to calculate the results
2. Random search : Instead of of calculating for all the combination in Random search CV, random samples are chosed and within given range.
3. Bayesian Optimization: A probabilistic technique is used to guide to the search of optimal hyperparameters.
