```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
```

```python
data= np.random.random((1000,784))
labels=np.random.randint(0,10,1000)
```

```python
df=pd.DataFrame(data)
df.sample(5)
```

```python
std=StandardScaler()
```

```python
standardized_data=std.fit_transform(data)
```

```python
standardized_data.shape
```

```python
tsne=TSNE(perplexity=50,n_components=5,method='exact') # method = exact which is exact algorithm instead of some fast trick,
                                                        # method exact is needed for n_componensts more than 3
```

```python
pc_data=tsne.fit_transform(standardized_data)
```

```python
new_df=pd.DataFrame(pc_data)
new_df.sample(5)
```
