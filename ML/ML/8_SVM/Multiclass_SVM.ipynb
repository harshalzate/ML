```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score

# Load the dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Create an SVM model with One-vs-Rest strategy
ovr_svm = OneVsRestClassifier(SVC(kernel='linear'))
ovr_svm.fit(X_train, y_train)

# Predict and evaluate the model
y_pred = ovr_svm.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'OvR SVM Accuracy: {accuracy}')
```

```python
from sklearn.multiclass import OneVsOneClassifier

# Create an SVM model with One-vs-One strategy
ovo_svm = OneVsOneClassifier(SVC(kernel='linear'))
ovo_svm.fit(X_train, y_train)

# Predict and evaluate the model
y_pred = ovo_svm.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'OvO SVM Accuracy: {accuracy}')
```

```python
# Create an SVM model directly
svc = SVC(kernel='linear', decision_function_shape='ovr')
svc.fit(X_train, y_train)

# Predict and evaluate the model
y_pred = svc.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f'SVC Direct OvR Accuracy: {accuracy}')
```
